```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pbl_dataset.csv")
df.head(50)
```

```python
df.shape
df.info()
df.describe()
df.columns
```

```python
df['Gender'] = df['Gender'].str.capitalize()
print(df)
```

```python
print("Null values:")
print(df.isnull().sum())
```

```python
print(df.nunique())
```

```python
df.hist(figsize=(12,10))
plt.show()
```

```python
print("LINE GRAPH")
plt.plot(df['Age'], df['Income'])
plt.title("Age vs Income")
plt.show()

plt.plot(df['Age'], df['LoanAmount'])
plt.title("Age vs Loan Amount")
plt.show()

plt.plot(df['Income'], df['LoanAmount'])
plt.title("Income vs Loan Amount")
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.show()
sns.countplot(x='Employment', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title("BOXPLOT")
sns.boxplot(data=df[['Income','LoanAmount','Savings']])
plt.show()
```

```python
Q1 = df['Age'].quantile(0.25)

Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1

df = df[(df['Age'] >= Q1 - 1.5*IQR) & (df['Age'] <= Q3 + 1.5*IQR)]

df
```

```python
plt.scatter(df.index, df['Age'])
plt.title("Outliers")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
from scipy import stats
z_scores = np.abs(stats.zscore(df['Income']))
df_clean = df[z_scores < 3]
print("Original Shape:", df.shape)
print("After Removing Outliers:", df_clean.shape)
```

```python
print("Mean:\n", df[['Income','LoanAmount','Savings']].mean())
print("\nMedian:\n", df[['Income','LoanAmount','Savings']].median())
print("\nMax:\n", df[['Income','LoanAmount','Savings']].max())
print("\nMin:\n", df[['Income','LoanAmount','Savings']].min())
print("\nStd Dev:\n", df[['Income','LoanAmount','Savings']].std())
```
